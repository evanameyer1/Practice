## **Matplotlib Overview and Practice**

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### **Scatter Plots**

```python
x = np.random.random(50) * 100
y = np.random.random(50) * 100

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.random(1000) * 100
y = np.random.random(1000) * 100

plt.scatter(x, y, c='r', s=125, marker='*', alpha=0.3)
plt.show()
```

#### **Line Plots**

```python
years = [2006 + x for x in range(16)]
weights = np.random.random(16) * 80

plt.plot(years, weights)
plt.show()
```

```python
years = [2006 + x for x in range(16)]
weights = np.random.random(16) * 80

plt.plot(years, weights, c='g', lw=5, linestyle=':')
plt.show()
```

#### **Bar Plots**

```python
x = ['C++', 'C#', 'Python', 'Java', 'SQL']
y = [np.random.randint(0, 200) for i in range(5)]

plt.bar(x, y)
plt.show()
```

```python
x = ['C++', 'C#', 'Python', 'Java', 'SQL']
y = [np.random.randint(0, 200) for i in range(5)]

plt.bar(x, y, color='violet', align='edge', width=0.5, edgecolor='purple')
plt.show()
```

#### **Hisograms**

```python
ages = np.random.normal(20, 1.5, 1000)

plt.hist(ages)
plt.show()
```

```python
ages = np.random.normal(20, 7.5, 10000)

plt.hist(ages, bins=25)
plt.show()
```

```python
ages = np.random.normal(20, 1.5, 1000)

plt.hist(ages, bins=[ages.min(), 17, 19, 21, ages.max()], cumulative=True)
plt.show()
```

#### **Pie Charts**

```python
langs = ['Python', 'C++', 'SQL', 'Java', 'JS']
votes = [np.random.randint(0, 75) for i in range(5)]

plt.pie(votes, labels=langs)
plt.show()
```

```python
langs = ['Python', 'C++', 'SQL', 'Java', 'JS']
votes = [np.random.randint(0, 75) for i in range(5)]
explodes = [0.2, 0, 0, 0, 0]

plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%", 
        pctdistance=1.5, startangle=45)
plt.show()
```

#### **Box Plots**

```python
heights = np.random.normal(172, 8, 300)

plt.boxplot(heights)
plt.show()
```

```python
first = np.linspace(0, 10, 25)
second = np.linspace(10, 200, 25)
third = np.linspace(200, 210, 25)
fourth = np.linspace(210, 230, 25)

data = np.concatenate((first, second, third, fourth))

plt.boxplot(data)
plt.show()
```

#### **Plot Customization**

```python
years = list(range(2014, 2021))
income = [np.random.randint(50, 80) for i in range(len(years))]

income_ticks = list(range(50, 81, 4))

plt.plot(years, income)
plt.title("Evan's Income", fontsize=20, fontname='Arial')
plt.xlabel('Year')
plt.ylabel('Annual Income in USD')
plt.yticks(income_ticks, [f'${x}k USD' for x in income_ticks])

plt.show()
```

```python
from matplotlib import style
```

```python
style.use('ggplot')

years = list(range(2014, 2021))
income = [np.random.randint(50, 80) for i in range(len(years))]

income_ticks = list(range(50, 81, 4))

plt.plot(years, income)
plt.title("Evan's Income", fontsize=20, fontname='Arial')
plt.xlabel('Year')
plt.ylabel('Annual Income in USD')
plt.yticks(income_ticks, [f'${x}k USD' for x in income_ticks])

plt.show()
```

#### **Legends and Multiple Plots**

```python
style.use('default')

stocka = [np.random.randint(90, 120) for i in range(10)]
stockb = [np.random.randint(90, 120) for i in range(10)]
stockc = [np.random.randint(90, 120) for i in range(10)]

plt.plot(stocka, label='Company1')
plt.plot(stockb, label='Company2')
plt.plot(stockc, label='Company3')
plt.legend(loc='lower right')

plt.show()
```

```python
langs = ['Python', 'C++', 'SQL', 'Java', 'JS']
votes = [np.random.randint(0, 75) for i in range(5)]

plt.pie(votes, labels=None)
plt.legend(labels=langs)
plt.show()
```

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sine Wave')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('Cosine Wave')

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title('Random Function')

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title('Log Function')

fig.suptitle('Four Plots')
plt.tight_layout()
plt.savefig('fourplots.png', dpi=300, transparent=False, bbox_inches='tight')
```

#### **Advanced Plotting**

```python
ax = plt.axes(projection='3d')

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title('3d Plot')
plt.show()
```

```python
ax = plt.axes(projection='3d')

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z)
ax.set_title('3d Plot')
plt.show()
```

```python
def fix_playernames():
    for i, player in enumerate(player_salaries['PLAYER']):
        if "," in player:
            player_salaries.loc[i, 'PLAYER'] = player.split(', ')[1] + ' ' + player.split(', ')[0]
        else:
            j = len(player) - 1
            while j > 0:
                if player[j].isupper():
                    if player[j-2:j] == "Mc":
                        j -= 2
                    elif player[j-1] == "-":
                        j -= 1
                    player_salaries.loc[i, 'PLAYER'] = player[:j] + " " + player[j:]
                    break
                j -= 1
```

```python
fix_playernames()
player_salaries
```

```python
def correct_player_names():
    final_df['PATCH'] = 'FIX'
    for name in player_salaries['PLAYER']:
        mask = final_df['PLAYER'].str.contains(name, case=False, na=False)
        final_df.loc[mask, 'PLAYER'] = name
        final_df.loc[mask, 'PATCH'] = 'PATCHED'
    final_df.reset_index(drop=True, inplace=True)
    final_df.dropna(inplace=True)
    return final_df[final_df['PLAYER'] != 'Empty']
```

```python
def set_correct_player_names():
    for i, player in enumerate(names_to_patch['PLAYER']):
        name_parts = player.split(' ', maxsplit=1)

        if len(name_parts) > 1:
            first_name = name_parts[0]
            last_name = name_parts[1]

            if last_name.startswith("D/ST"):
                last_name = "D/ST"
                
            else:
                
                if "Jr." not in last_name and len(last_name) > 0:
                    count = 0
                    j = len(last_name) - 1
                    while j > 0:
                        if last_name[j].isupper():
                            if count == 1: 
                                last_name = last_name[:j]
                            else:
                                count += 1
                        j -= 1
                elif "Jr." in last_name:
                    try:
                        last_name = last_name[:last_name.index('.') + 1]
                    except ValueError:
                        pass

            names_to_patch.at[i, 'PLAYER'] = f"{first_name} {last_name}"
            
    names_to_patch.reset_index(drop=True, inplace=True)
```

```python
set_correct_player_names()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
mpg.tail()
mpg.describe()
```

```python
sns.set_theme(style='white')
```

```python
df1 = sns.relplot(data=mpg,
                x='horsepower', 
                y='mpg', 
                hue='origin',
                size='weight',
                sizes=(40, 200),
                alpha=0.5, 
                kind='scatter')

df1.fig.suptitle('Scatterplot with varying point sizes and hues', fontsize=20)
df1.fig.subplots_adjust(top=0.9);

plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme(style="whitegrid")
```

```python
sns.violinplot(data=tips,
               x='day',
               y='total_bill',
               hue='smoker',
               split=True,
               inner='quart',
               linewidth=1,
               palette={"Yes": "b", "No": ".85"});
plt.show()
```

```python
sns.jointplot(data=tips,
              x="total_bill", 
              y="tip",
              kind="reg", 
              truncate=False,
              xlim=(0, 60), 
              ylim=(0, 12),
              color="m", 
              height=7)
plt.show()
```

```python
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import re
from selenium.webdriver.support.ui import Select
```

```python
def main(driver, account):
    driver.maximize_window()
    driver.implicitly_wait(10)
    driver.get(account)
    select_us(driver)
    df = gather_data(driver)
    print('data gathered')
    return df
```

```python
def select_us(driver):
    select = Select(driver.find_element(By.CSS_SELECTOR, "#Region"))
    select.select_by_value('.united-states')
```

```python
def gather_data(driver):
    company_names = []
    company_websites = []
    company_oneliners = []
    company_industries = []
    count = 1
    while True:
        try: 
            #for each company on the website
            if driver.find_element(By.CSS_SELECTOR, f"#elitemix > div:nth-child({count}) > div.elite__item-box"):
                #grab the website url
                link = driver.find_element(By.CSS_SELECTOR, f"#elitemix > div:nth-child({count}) > div.elite__item-box > a")
                company_websites.append(link.get_attribute("href"))
                
                #grab the company name
                company_name = driver.find_element(By.CSS_SELECTOR, f"#elitemix > div:nth-child({count}) > div.elite__item-box > div.elite__collection-item-content > h4").text
                company_names.append(company_name)
            
                #grab the company oneliner
                company_oneliner = driver.find_element(By.CSS_SELECTOR, f"#elitemix > div:nth-child({count}) > div.elite__item-box > div.elite__collection-item-content > div > div:nth-child(2)").text
                company_oneliners.append(company_oneliner)
                
                #grab the company industries
                company_industry = driver.find_element(By.CSS_SELECTOR, f"#elitemix > div:nth-child({count}) > div.elite__item-box > div.elite__collection-item-content > div > div.elite__item-category-tag").text
                company_industries.append(company_industry)
                
                count += 1
        except:
            break
            
    df = pd.DataFrame(list(zip(company_names, company_websites, company_oneliners, company_industries)), columns =['company', 'website', 'overview', 'industry']) 
    return df
```

```python
df = main(webdriver.Chrome(executable_path="path"), 'https://www.asugsvsummit.com/gsv-cup-elite-200')
```
